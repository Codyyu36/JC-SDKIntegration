```python
def make_image_no_noise(base_level, prop=0.21):
    os.makedirs('patterns/white'+'/',exist_ok=True)
    pattern = black
    l, w = LENGTH * prop, WIDTH * prop *1.65
    for row in range(round(LENGTH/2-l/2),round(LENGTH/2+l/2)):
        for col in range(round(WIDTH/2-w/2),round(WIDTH/2+w/2)):

            pattern[row, col] = [base_level, base_level, base_level]


    pattern = Image.fromarray(pattern)
    pattern.save('patterns/white'+ "/" + str(base_level)  + "_" + str("good") + '.bmp')

    return pattern
```

```python
def make_image(base_level, variation, prop=0.21):
    os.makedirs('patterns/white'+'/',exist_ok=True)
    pattern = black
    l, w = LENGTH * prop, WIDTH * prop *1.65
    for row in range(round(LENGTH/2-l/2),round(LENGTH/2+l/2)):
        for col in range(round(WIDTH/2-w/2),round(WIDTH/2+w/2)):
            r = random.randint(base_level - variation, base_level + variation)
            g = random.randint(base_level - variation, base_level + variation)
            b = random.randint(base_level - variation, base_level + variation)
            pattern[row, col] = [r, g, b]

    for row in range(round(LENGTH/2-l/2),round(LENGTH/2+l/2)):
        for col in (round(WIDTH/2-w/2),round(WIDTH/2+w/2)):
            pattern[row, col] = [base_level, base_level, base_level]

    for row in (round(LENGTH/2-l/2),round(LENGTH/2+l/2)):
        for col in range(round(WIDTH/2-w/2),round(WIDTH/2+w/2)):
            pattern[row, col] = [base_level, base_level, base_level]

    pattern = Image.fromarray(pattern)
    pattern.save('patterns/white'+ "/" + str(base_level) + "_" + str(variation) + "_" + str("noisy") + '.bmp')

    return pattern
```

```python
def make_image_edges(base_level, prop=0.21):
    os.makedirs('patterns/white'+'/',exist_ok=True)
    pattern = black
    l, w = LENGTH * prop, WIDTH * prop *1.65
    for row in range(round(LENGTH/2-l/2),round(LENGTH/2+l/2)):
        for col in (round(WIDTH/2-w/2),round(WIDTH/2+w/2)):
            pattern[row, col] = [base_level, base_level, base_level]

    for row in (round(LENGTH/2-l/2),round(LENGTH/2+l/2)):
        for col in range(round(WIDTH/2-w/2),round(WIDTH/2+w/2)):
            pattern[row, col] = [base_level, base_level, base_level]

    
    pattern = Image.fromarray(pattern)
    pattern.save('patterns/white'+ "/" + str(base_level) + "_" + str(base_level) + "_" + str("edges") + '.bmp')

    return pattern
```
